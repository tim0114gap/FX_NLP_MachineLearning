```python
import urllib3
from bs4 import BeautifulSoup
import certifi
```

```python
#アクセスするURL
url = "https://www.google.com/search?rlz=1C6GCEA_enKR868KR868&sxsrf=ACYBGNQoU_oqPa3Z2AM7zVFXIAxh8Sy4Ug:1571052415874&q=USD/JPY+news&tbm=nws&source=univ&tbo=u&sa=X&ved=2ahUKEwjAva-p0pvlAhVsHKYKHbE2Ar0Qt8YBKAF6BAgFEA4&biw=1440&bih=789"
```

```python
# httpsの証明書検証を実行している
http = urllib3.PoolManager()
r = http.request('GET', url)
```

```python
soup = BeautifulSoup(r.data, 'html.parser')
soup2 = BeautifulSoup(r.data, 'html.parser')
```

```python
# タイトル取得
title_tag = soup.title
print(title_tag)

# 要素の文字列を取得
title = title_tag.string
print(title)
```

```python
span = soup.find_all(class_="st")
print(span)
```

```python
lst = ['<div class="st">',"</div>","<b>","</b>","...","(",")"]
# corpus = []
# [corpus.append(list(i)) for i in span]

word = []
for text in span:
    text = str(text)
    for x in lst:
        text = text.replace(x,"")
        text = text.lower()
    word.append(text)
```

```python
word
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = DataFrame(data={'word':word})
df.head()
```

```python
pre_df = pd.read_csv('corpusUSD_YPN_news.csv')
del pre_df['Unnamed: 0']
pre_df.head()
```

```python
new = []
for x in df.word:
    if x not in pre_df.word:
        new.append(x)

df2 = DataFrame({'word':new})
df2
```

```python
pre_df = pre_df.append(df2, ignore_index=True)
pre_df
```

```python
pre_df.to_csv("corpusUSD_YPN_news.csv")
```
