```python
import numpy as np
import pandas as pd
import preprocessor as p
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
pd.options.display.float_format = '{:0.2f}'.format
```

```python
coupas = []
path = "news.txt"
with open(path) as f:
    s = f.read()
s
```

```python
def remove_punctuation(text):
    no_punc = "".join([x for x in text if x not in string.punctuation])
    return no_punc
```

```python
s = remove_punctuation(s)
s
```

```python
# instantiate tokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
s = tokenizer.tokenize(s.lower())
s
```

```python
def remove_stopwords(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words
```

```python
s = remove_stopwords(s)
s
```

```python
# instantiate lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(x) for x in text]
    return lem_text
```

```python
s = word_lemmatizer(s)
s
```

```python
# instantiate stemmer
stemmer = PorterStemmer()
```

```python
def word_stemmer(text):
    stem_txt = " ".join([stemmer.stem(x) for x in text])
    return stem_txt
```

```python
s = word_stemmer(s)
s
```

```python
coupas = []
coupas.append(s)
```

### bag of word
https://blog.amedama.jp/entry/tf-idf

```python
# 単語の数をカウントする
count_vectorizer = CountVectorizer()
bow = count_vectorizer.fit_transform(coupas).toarray()
bow
```

```python
print('--- BoW (Bag of Words) ---')
df1 = pd.DataFrame(bow,
                  columns=count_vectorizer.get_feature_names())
df1.head()
```

```python
topWord = df1.iloc[0][df1.iloc[0] >= 10]
topWordf = pd.DataFrame({"counts":topWord})
topWordf
```

```python
topWordf.to_csv("bag_of_words.csv")
```
